```python
import requests
import json
from datetime import datetime
import pandas as pd
```

```python
# Datos
url = 'http://api.citybik.es/v2/networks/citi-bike-nyc'
```

```python
resp = requests.get(url)
json_resp = resp.json()
```

```python
stations = json_resp.get('network').get('stations')
```

```python
bikes = [(station['name'], station['empty_slots'], station['free_bikes'], station['timestamp'])
 for station in stations]


pd.DataFrame(bikes, columns = ['station', 'empty_slots', 'free_bikes', 'timestamp'])
```
